```python
!pip install faster-whisper
from faster_whisper import WhisperModel

# Configuration
WHISPER_MODEL = "base"  # Options: tiny, base, small, medium, large
AUDIO_FILE = "extracted_audio.mp3"  # Replace with your audio file name

# Transcribe audio
print(f"Loading faster-whisper model: {WHISPER_MODEL}")
model = WhisperModel(WHISPER_MODEL, device="cuda", compute_type="float16")

print("Running transcription with faster-whisper...")
segments_generator, info = model.transcribe(AUDIO_FILE, word_timestamps=True)

# Function to format time for SRT
def format_srt_time(seconds):
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    millis = int((seconds % 1) * 1000)
    return f"{hours:02d}:{minutes:02d}:{secs:02d},{millis:03d}"

# Create SRT content
srt_content = ""
for i, segment in enumerate(segments_generator, 1):
    start_time = format_srt_time(segment.start)
    end_time = format_srt_time(segment.end)
    text = segment.text.strip()

    srt_content += f"{i}\n"
    srt_content += f"{start_time} --> {end_time}\n"
    srt_content += f"{text}\n\n"

# Save transcription to SRT file
with open("transcription.srt", "w", encoding="utf-8") as f:
    f.write(srt_content)

print("Transcription saved to transcription.srt. Download this file and provide its path to the local script.")
```
